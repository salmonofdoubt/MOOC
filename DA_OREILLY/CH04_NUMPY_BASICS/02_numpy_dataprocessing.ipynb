# Data Processing using Arrays

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Slightly bigger 1D array
points = np.arange(-5, 5, 0.01) 
xs, ys = np.meshgrid(points, points)
```

```python
# Explore sqrt(xs^2 + ys^2)
z = np.sqrt(xs**2 + ys**2)
print(z)
```

```python
# using matplotlib
plt.imshow(z, cmap=plt.cm.gist_rainbow); plt.colorbar()
plt.title("Image of $\sqrt{x^2 + y^2}$ for a grid of values")
```

## Conditional Logic as Array operations

```python
xarr = np.arange(1.1, 1.6, 0.1)
yarr = np.arange(2.1, 2.6, 0.1)
cond = np.array([True, False, True, True, False])
print(xarr)
print(yarr)
```

```python
result = []
for x, y, c in zip(xarr, yarr, cond):
    if c:
        result.append(x)
    else:
        result.append(y)

print(result)

# same as list comprehension:
result2 = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]

print(result2)

# same using np.where (where cond is True)
result3 = np.where(cond, xarr, yarr)
print(result3)
```

```python
arr = np.random.randn(4,4)
# --------------> (cond, if True, if False)
result4 = np.where(arr > 0, 1, -1)
result4 

# one line
np.where(np.random.randn(4,4) > 0, 1, -1)
```

## Mathematical and Statistical Methods

```python
arr = np.random.randn(5,4)
print(arr)
#arr has now inherited methods 
arr.mean() # all values
#np.mean(arr)
```

```python
arr2 = np.arange(0,6)
print(arr2)
arr2.mean() # there is only one axix
```

```python
arr3 = np.array([[[1,1,1],[2,2,2]],[[3,3,3],[4,4,4]]])
print(arr3)
arr3.mean(axis=0) # data along the 0 axis y sky, 'axis' not necessary
```

```python
arr3.mean(1) # data along the first axis x
```

```python
arr3.mean(2) # data along the second axis z
```
