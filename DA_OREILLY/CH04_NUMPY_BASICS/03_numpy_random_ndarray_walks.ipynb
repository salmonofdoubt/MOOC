# Random Walks around a NDArray

```python
# Classic solution
import random
import matplotlib.pyplot as plt
position = 0
walk = [position]
steps = 100

for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)

print('min:', min(walk), 'max:', max(walk))

# Randomness does not average out :)
# 'walk' is the cumulative sum of the random steps
plt.plot(walk) 
plt.ylabel('Position y after x steps')
plt.show()
```

```python
# Use numpy for flipping coin
import numpy as np
nsteps = 100
draws = np.random.randint(0,2, size=nsteps) # between 0 and 1 
print('draws:', draws)
steps = np.where(draws > 0, 1, -1)
print('steps:', steps)
walk = steps.cumsum()
print('walk:', walk)

print('min:', walk.min(), 'max:', walk.max())

plt.plot(walk) 
plt.ylabel('Position y after x steps')
plt.show()

ten_steps = (np.abs(walk) >= 10).argmax()
print('First time 10 steps away:', ten_steps)
```
